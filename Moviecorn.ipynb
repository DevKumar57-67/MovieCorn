```python
# Take user input from user then return the recommended movie to watch

import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the movie data (assuming it's in the same directory or provide the correct path)
movies = pd.read_csv("/content/netflix_titles.csv")

# Drop unnecessary columns (same as your original code)
movies = movies.drop(['show_id', 'director', 'cast', 'country', 'date_added', 'rating', 'release_year', 'duration', 'listed_in'], axis=1)

# Create TF-IDF matrix (same as your original code)
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['description'])
cos_theta = cosine_similarity(tfidf_matrix, tfidf_matrix)


# Function to recommend movies based on description (same as your original code)
def recommend_movies_by_description(movie_title, num_recommendations=5):
    """
    Recommends movies based on the similarity of their descriptions.

    Args:
        movie_title: The title of the movie for which to find recommendations.
        num_recommendations: The number of recommendations to return.

    Returns:
        A list of recommended movie titles.
    """

    try:
        idx = movies[movies['title'] == movie_title].index[0]
        sim_scores = list(enumerate(cos_theta[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        sim_scores = sim_scores[1:num_recommendations + 1]
        movie_indices = [i[0] for i in sim_scores]
        return movies['title'].iloc[movie_indices].tolist()

    except IndexError:
        return "Movie not found in the dataset."

# Get user input
user_input = input("Enter the previous movie which you liked to get recommended the new one: ")

# Get recommendations
recommendations = recommend_movies_by_description(user_input)

# Print recommendations
if isinstance(recommendations, list):
    print("Recommended movies:")
    for movie in recommendations:
        print(movie)
else:
    print(recommendations) # Print the "Movie not found" message
```
